Q6 Run basic calculations to obtain the results in the table below

Create a summary data frame to hold the results

Sort the total purchase value column in descending order

Optional: give the displayed data cleaner formatting

Display a preview of the summary data frame

```python
#create df group by SN
Top_by_sn  = all_data.groupby(['SN'])
```

```python
#create df for Purchase Count,Average Purchase Price & Total Purchase Value

Topbysn_count = Top_by_sn ["Price"].count()
#print(sn_price_count.head())
Topbysn_count_mean = Top_by_sn ["Price"].mean()
#print(sn_price_mean.head())
Topbysn_TotPur = Top_by_sn ["Price"].sum()
#print(sn_price_sum.head())
```

```python
print(Topbysn_count)
print(Topbysn_count_mean)
print(Topbysn_TotPur)
```

```python
#Create summary data frame
BySN_TopSpenders = pd.DataFrame({
                   "Purchase Count": Topbysn_count, 
    "Average Purchase Price":Topbysn_count_mean, 
    "Total Purchase Value": Topbysn_TotPur
}, columns=["Purchase Count", "Average Purchase Price", "Total Purchase Value"])   
```

```python
BySN_TopSpenders
```

```python
#Display and Sort
BySN_TopSpenders = BySN_TopSpenders.sort_values(['Total Purchase Value'], ascending=False).head()
BySN_TopSpenders
```

```python
#create df group by Item ID, Item Name
Top_by_ItemID2  = all_data.groupby(['Item ID','Item Name'])
```

```python
#create df for Purchase Count,Average Item Price & Total Purchase Value

PopItem_count = Top_by_ItemID2 ["Price"].count()

PopItem_price = Top_by_ItemID2 ["Price"].mean()

PopItem_TotPur = Top_by_ItemID2 ["Price"].sum()
```

```python
BySN_Popitems.head()
```
